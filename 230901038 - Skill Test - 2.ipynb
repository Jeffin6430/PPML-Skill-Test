```python
"""1. Signal Classification Using Machine Learning
Problem Statement: You have a dataset of electrical signals, where each signal
represents either a &quot;sine wave&quot; or a &quot;square wave.&quot; The dataset contains the following
features for each signal: - Amplitude (in volts) , Frequency (in Hz) , Duration (in
seconds) , Peak-to-Peak Voltage (in volts) .
Your task is to develop a machine learning model that can classify each signal as
either &quot;Sine&quot; or &quot;Square&quot; based on these features using KNN. Encode labels (Sine →
0, Square → 1)"""
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("PPML1.csv")
x=df['Amplitude']
y=df['Duration']
cl=df['Label'].map({'Sine':0,'Square':1})
data=list(zip(x,y))
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(data,cl)
nx=1.7
ny=1.4
np=[(nx,ny)]
prediction=knn.predict(np)
plt.scatter(x,y,c=cl)
plt.scatter(nx,ny,marker='*')
plt.xlabel('Amplitude')
plt.ylabel('Duration')
plt.title('Sine vs Square')
plt.show()
```

```python
"""2. Write a program to classify a component is defective or not based on Voltage and
Current using KNN. Encode labels (Working → 1, Defective → 1)"""
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
df=pd.read_csv("PPML2.csv")
x=df['Voltage (V)']
y=df['Current(A)']
cl=df['Status'].map({'Defective':0,'Working':1})
d=list(zip(x,y))
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(d,cl)
nx=5
ny=0.6
np=[(nx,ny)]
prediction=knn.predict(np)
plt.scatter(x,y,c=cl,cmap='coolwarm')
pc='red' if prediction[0]==1 else 'blue'
plt.scatter(nx,ny,c=pc,marker='*')
plt.xlabel('Voltage (V)')
plt.ylabel('Current(A)')
plt.title('Component is Defective or Working')
plt.show()
```

```python
"""3. Write a program to Predict Final Exam Scores based on features like study hours
and Final exam score using Linear Regression."""
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
df=pd.read_csv("PPML3.csv")
x=df['Study Hours']
y=df['Final Exam Score']
slope,intercept,r,p,std_err=stats.linregress(x,y)
def f(x):
  return slope*x+intercept
mymodel=list(map(f,x))
plt.scatter(x,y,color='red',label='Actual Exam Scores')
plt.plot(x,mymodel,color='purple',label='Linear Regression')
plt.scatter(8.85,f(8.85),color='black',label='Predicted Exam Score',marker='x')
plt.legend()
plt.title('Study Hours and Final Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Final Exam Score')
plt.show()
```

```python
"""4. Write a program to build a machine learning linear regression model to predict
the price of a house based on its features. You are given a dataset containing
information of Year and price."""
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
df=pd.read_csv("PPML4.csv")
x=df['Year']
y=df['Price']
slope,intercept,r,p,std_err=stats.linregress(x,y)
def f(x):
  return slope*x+intercept
mymodel=list(map(f,x))
plt.scatter(x,y,color='red',label='Actual Price')
plt.plot(x,mymodel,color='blue',label='Linear Regression')
plt.scatter(2021,f(2021),color='brown',marker='X',label='Predicted Price')
plt.legend()
plt.title('Price Prediction over the years')
plt.xlabel('Year')
plt.ylabel('House Price')
plt.grid(True)
plt.show()
```

```python
"""5. Write a program to develop a decision tree classifier to predict signal quality (Good or
Bad) based on transmitter, signal strength, frequency."""
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
df=pd.read_csv("PPML5.csv")
d={'T1':0,'T2':1,'T3':2}
df['Transmitter']=df['Transmitter'].map(d)
d={'Good':0,'Bad':1}
df['Signal Quality']=df['Signal Quality'].map(d)
features=['Transmitter','Signal Strength','Frequency']
x=df[features]
y=df['Signal Quality']
dectree=DecisionTreeClassifier()
dectree.fit(x,y)
tree.plot_tree(dectree,feature_names=features)
plt.show()
```

```python
"""6. Write a program to Generate a synthetic dataset representing electronic components based on their
voltage and current characteristics using Kmeans Clustering.
The dataset contains two clusters based on Voltage and Current.
The dataset contains two clusters based on Voltage and Current.
 Cluster 0 corresponds to low voltage and current values.
 Cluster 1 corresponds to higher voltage and current values."""
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
df=pd.read_csv("PPML6.csv")
x=df['Voltage']
y=df['Current']
d=list(zip(x,y))
kmeans=KMeans(n_clusters=2)
kmeans.fit(d)
plt.scatter(x,y,c=kmeans.labels_)
plt.xlabel('Voltage')
plt.ylabel('Current')
plt.title('VI characteristics')
plt.show()
```
