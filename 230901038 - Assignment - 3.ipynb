```python
"""2. Write a Python script using pandas for Student Performance Analysis:
Problem Statement:
You are given a CSV file named "student_scores.csv" containing scores of students in
various subjects.
Perform the following tasks:
a. Display the summary statistics for each subject,
b. Calculate the average score for each student,
c. Identify students who scored below 60 in more than two subjects."""
import pandas as pd
df=pd.read_csv("student scores.csv")
print("Original data:\n",df)
print("\nSummary statistics for each subject:")
print(df[['Maths','Physics','Chemistry']].describe())
df['Average']=df[['Maths','Physics','Chemistry']].mean(axis=1)
print('\n',df[['Student ID','Name','Average']])
below=df[['Maths','Physics','Chemistry']]<60
df['b']=below.sum(axis=1)>2
print("\nStudents below 60 marks in more than 2 subjects:")
c=0
for i in range(len(df)):
  if df['b'][i]==True:
    print(df['Student ID'][i])
    c+=1
if c==0:
  print("     No student scored below 60 in more than 2 subjects")
```

```python
"""3. Write a Python script for Flight Data Analysis:
Problem Statement:
You have a CSV file named &#39;flight_data.csv&#39; containing flight information such as origin,
destination, and departure time.
Perform the following tasks:
a. Display the summary statistics for departure delays,
b. Calculate the average delay for each airline,
c. Identify the most common departure and arrival destinations,
d. Visualize the distribution of departure delays using a box plot."""
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
df=pd.read_csv("flight data.csv")
print("Original dataframe:",df)
print("\nSummary statistics for departure delays:\n",df['Departure Delay'].describe())
avg=df.groupby('Airline').agg({'Departure Delay':'mean'})
print("\nAverage Delay for each airline:\n",avg)
print("\nMost common origin:",df['Origin'].mode()[0])
print("Most common Destination:",df['Destination'].mode()[0])
print("\n")
sb.boxplot(x='Airline',y='Departure Delay',color='blue',data=df)
plt.title('Departure Delay by Airline')
plt.show()
```

```python
"""5. Write a program to work with time series data using NumPy and pandas
Problem Statement: A time series dataset representing daily temperatures over a year.
You will perform some basic analysis and visualization using NumPy, pandas, and
matplotlib."""
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
date=pd.date_range(dt.date(2025,1,1),periods=365)
temp=np.random.randint(15,36,size=365)
df=pd.DataFrame({'Date':date,'Temperature':temp})
print("Original Data:\n",df)
print("\nBasic Analysis:\n",df['Temperature'].describe())
df.set_index('Date', inplace=True)
monthly_avg = df.resample('ME').mean()
print("\nMonthly Average Temperatures:")
print(monthly_avg)
plt.plot(monthly_avg.index, monthly_avg['Temperature'], label='Monthly Avg Temp',color='red')
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Monthly Average Temperatures')
plt.show()
```

```python
"""6. Write a program to simulate and analyze daily stock prices over one year and generate
the stock prices, calculate some basic metrics, and visualize the data with additional
features like moving averages and visualize the same."""
import datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
date=pd.date_range(dt.date(2025,1,1),periods=365)
stock=np.random.randint(80,115,size=365)
df=pd.DataFrame({'Date':date,'Stock Price':stock})
print("Original DataFrame:\n",df)
print("\nBasic Metrics:",df['Stock Price'].describe())
df['50-day MA']=df['Stock Price'].rolling(window=50).mean()
df['200-day MA']=df['Stock Price'].rolling(window=200).mean()
fig=plt.figure(figsize=(12,4))
plt.plot(df.index,df['Stock Price'],label='Stock Price',color='blue')
plt.plot(df.index,df['50-day MA'],label='50-day Moving Average',color='red', linestyle='-.')
plt.plot(df.index,df['200-day MA'],label='200-day Moving Average',color='green')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Simulated Stock Price and Moving Averages')
plt.legend()
plt.show()
```
